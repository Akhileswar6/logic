```python
n = int(input("Enter number of rows: "))

# Upper half (inverted pyramid)
for i in range(n, 0, -1):
    print(" " * (n - i), end="")   # spaces
    print("* " * i)                # stars

# Lower half (pyramid)
for i in range(2, n + 1):
    print(" " * (n - i), end="")   # spaces
    print("* " * i)                # stars



```

```python
n = int(input("Enter number of rows: "))

# Upper half
for i in range(1, n + 1):
    # Left stars
    print("* " * i, end="")
    # Spaces in middle
    print("  " * (n - i) *2, end="")
    # Right stars
    print("* " * i)

# Lower half
for i in range(n, 0, -1):
    # Left stars
    print("* " * i, end="")
    # Spaces in middle
    print("  " * (n - i) * 2, end="")
    # Right stars
    print("* " * i)
```
